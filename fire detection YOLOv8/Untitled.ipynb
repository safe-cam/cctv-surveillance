## using web came

```python
from ultralytics import YOLO
```

```python
file = r"runs\detect\train9\weights\best.pt"
model = YOLO(file)
```

```python
import cv2

video_path = "video.mp4"
cap = cv2.VideoCapture(video_path)

frame_width = 640
frame_height = 480

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (frame_width, frame_height))

    results = model.predict(source=frame, conf=0.23, save=False)

    for r in results:
        annotated_frame = r.plot()

        cv2.imshow("YOLOv8 Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
